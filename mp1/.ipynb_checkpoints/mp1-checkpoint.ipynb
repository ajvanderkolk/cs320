```python
def smart_multiply(nums) -> int:
    product = 1
    for num in nums:
        if product > 100:
            break
        if not 0 < num:
            continue
        else:
            product *= num
    return product
```

```python
#Q7
smart_multiply(nums)
```

```python
# convert to int helper function
def convert_to_int(value):
    try:
        return int(value)
    except ValueError:
        print("Could not convert string to int.")
    except Exception:
        print("Error with your input argument.")
    return None
```

```python
from subprocess import check_output
# helper function (turn git branch into list)
def gbl(cwdir, comm='branch'):
    return str(check_output(['git', comm], cwd=cwdir), 
          encoding='utf-8').split('\n')
```

```python
import numpy as np
import time
def rand_df(rows):
    return pd.DataFrame(np.random.randint(10, size=(rows, 5)), 
                        columns=['A','B','C','D','E'],
                        index=[f'r{i}' for i in range(1, rows+1)])
```

```python
#Q32
times_df = pd.DataFrame(dtype=float)

for num in range(1000, 7000, 1000):
    df = rand_df(num)
    
    # iterrows
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    for row in df.iterrows():
        pass
    t1 = time.time()
    times_df.at[num, 'iterrows'] = t1 - t0
    
    # itertuples
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    for row in df.itertuples():
        pass
    t1 = time.time()
    times_df.at[num, 'itertuples'] = t1 - t0

times_df.plot.line(xlabel = 'Number of Rows',
                   ylabel = 'Milliseconds',
                   title = 'Benchmarking: iterrows vs. itertuples')
```

```python
#Q33
times_df = pd.DataFrame(dtype=float)
for num in range(1000, 7000, 1000):
    df = rand_df(num)
    
    # .loc
    total = 0
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    for idx in df.index:
        for col in df.columns:
            total += df.loc[idx, col]
    t1 = time.time()
    times_df.at[num, '.loc'] = t1 - t0
    
    # .at
    total = 0
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    for idx in df.index:
        for col in df.columns:
            total += df.at[idx, col]
    t1 = time.time()
    times_df.at[num, '.at'] = t1 - t0

times_df.plot.line(xlabel = 'Number of Rows',
                   ylabel = 'Milliseconds',
                   title = 'Benchmarking: .loc vs. .at')
```

```python
# helper function for question 34
def laugh(x):
    return "qwerty" * x
```

```python
#Q34
times_df = pd.DataFrame(dtype=float)
for num in range(1000, 21000, 1000):
    df = rand_df(num)
    
    # apply
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    result = df["A"].apply(laugh).tolist()
    t1 = time.time()
    times_df.at[num, 'apply'] = t1 - t0
    
    # loop
    t0 = time.time()
        # HOW LONG DOES THIS TAKE TO RUN, FOR DATAFRAMES WITH DIFFERENT NUMBERS OF ROWS?
    result = []
    for val in df["A"]:
        result.append(laugh(val))
    t1 = time.time()
    times_df.at[num, 'loop'] = t1 - t0

times_df.plot.line(xlabel = 'Number of Rows',
                   ylabel = 'Milliseconds',
                   title = 'Benchmarking: apply vs. loop')
```
